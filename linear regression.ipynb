```python
# Step 1: Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Step 2: Load the Dataset
# For example, you can load the dataset using pandas.
# Replace this line with your actual dataset path.
data = {
    'Area': [2600, 3000, 3200, 3600, 4000],
    'Bedrooms': [3, 4, 3, 3, 5],
    'Price': [550000, 565000, 610000, 680000, 725000]
}
df = pd.DataFrame(data)

# Step 3: Preprocess the Data
# Define the features (X) and target (y)
X = df[['Area', 'Bedrooms']]  # Features: Area and Bedrooms
y = df['Price']  # Target: Price

# Step 4: Split Data into Train-Test Sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)


# Step 5: Fit the Linear Regression Model
model = LinearRegression()
model.fit(X_train, y_train)

# Step 6: Make Predictions
y_pred = model.predict(X_test)

# Step 7: Evaluate the Model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print Model Evaluation Metrics
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared Score (RÂ²): {r2}")

# Step 8: Plot the Regression Line (for Area vs Price)
plt.figure(figsize=(10,6))

# Scatter plot of the actual data points
plt.scatter(df['Area'], df['Price'], color='blue', label='Actual Data')

# Plot the regression line for Area vs Price
plt.plot(df['Area'], model.predict(df[['Area', 'Bedrooms']]), color='red', label='Regression Line')

# Add labels and title
plt.title('Price vs Area (with Bedrooms)')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price')
plt.legend()
plt.show()

# Step 9: Interpret Model Coefficients
print("Model Coefficients:")
print(f"Area Coefficient: {model.coef_[0]}")
print(f"Bedrooms Coefficient: {model.coef_[1]}")
print(f"Intercept: {model.intercept_}")
```
